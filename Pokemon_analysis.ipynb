### Description about the dataset

The attributes in the dataset are as follows:

- Pokedex entry number of the Pokemon
pokedex is an  catalog that provide information regarding the various species of Pokémon featured in the Pokémon video game, anime and manga series.)

- Name = name of the Pokemon

- Type 1 = each Pokemon has a type, this determines weakness/resistance to attacks [referred to as the primary type]

- Type 2 = some Pokemon are dual type and have 2 [referred to as the secondary type]

- HP = hit points, or health, defines how much damage a Pokemon can withstand before fainting

- Attack =  the base modifier for normal attacks

- Defense = the base damage resistance against normal attacks

- Sp. Atk = special attack, the base modifier for special attacks

- Sp. Def = the base damage resistance against special attacks

- Speed = determines which Pokemon attacks first each round

- Generation = refers to which grouping/game series the Pokemon was released in

- Legendary = a boolean that identifies whether the Pokemon is legendary

### Feature engineering

#### Creating a new feature named Total_power

This column has been created by summing the columns such as HP, Attack, Defense, Sp.Attack , Sp. Defense and Speed.

```python
data["Total_power"]=data["HP"]+data["Attack"]+data["Defense"]+data["Sp. Atk"]+data["Sp. Def"]+data["Speed"]
data.head()
```

#### Creating a new feature named total attack
This column has been created by summing two columns namely Attack and Sp.Attack

```python
data["Total_attack"]=data["Attack"]+data["Sp. Atk"]
data.head()
```

#### Creating a new feature named total defense
This column has been created by summing two columns namely Defense and Sp.Defense

```python
data["Total_defense"]=data["Defense"]+data["Sp. Def"]
data.head()
```

### Exploratory Data Analysis (EDA)

#### How many types of pokemon exist in this catelogue and what are there names?

```python
data["Type 1"].nunique()
```

```python
data["Type 1"].unique()
```

##### Observation
There are 18 types of pokemons.

#### Which type of pokemon has highest and lowest instances?

```python
sns.countplot(x="Type 1",data=data)
plt.xticks(rotation=90);
plt.show()
```

##### Observation
The water type pokemon are the most prevalent and the flying type are the least.

#### Which generation has more pokemon?

```python
data["Generation"].value_counts()
```

```python
sns.countplot(x="Generation",data=data)
plt.show()
```

#### Who is the strongest pokemon? Also find out its generation and type.

```python
data.sort_values(by="Total_power",ascending=False).head(1)[["Name","Generation","Type 1","Total_power"]]
```

##### Observation
RayquazaMega Rayquaza was found to be the strongest pokemon in terms of total power. It belongs to generation 3 and type 1.

#### Who is the weakest pokemon? Also find out its generation and type.

```python
data.sort_values(by="Total_power",ascending=False).tail(1)[["Name","Generation","Type 1","Total_power"]]
```

##### Observation
Sunkern was found to be the weakest pokemon in terms of total power. It belongs to generation 2 and type 1.

#### What is the generation wise category count?

```python
data.groupby(["Generation"])["Type 1"].value_counts()
```

##### Observation
Generation 1 to 3 has water type as the highest number of pokemon and generation 4 and 5 has normal type pokemon has highest.
Only the last generation has ghost type pokemon as the highest.
