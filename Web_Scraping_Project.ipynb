```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

* Requests library is used to send HTTP requests to a specified URL.

<div style="border-radius:10px; border:black solid; padding: 15px; background-color: purple; font-size:80%; text-align:left">
<p style="font-family:Georgia; font-weight:bold; letter-spacing: 2px; color:white; font-size:200%; text-align:center;padding: 0px;"> Step 2 - Getting response</p>

```python
url='https://www.flipkart.com/search?q=realme+poco+m4+pro&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_11_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_11_na_na_na&as-pos=2&as-type=RECENT&suggestionId=realme+poco+m4+pro%7CMobiles&requestId=80c5cdf1-8907-427d-8ec7-e6078ff99817&as-searchtext=realme%20poco'
response=requests.get(url)
print(response)
```

<div style="border-radius:10px; border:#DEB887 solid; padding: 15px; background-color: #FFFAF0; font-size:100%; text-align:left">

<h3 align="left"><font color='#DEB887'>ðŸ’¡ Note:</font></h3>

Response 200 implies that request connection to the website is successful.

<div style="border-radius:10px; border:black solid; padding: 15px; background-color: purple; font-size:80%; text-align:left">
<p style="font-family:Georgia; font-weight:bold; letter-spacing: 2px; color:white; font-size:200%; text-align:center;padding: 0px;"> Step 3 - Getting the html content</p>

```python
htmlcontent=response.text
print(htmlcontent)
```

```python
soup=BeautifulSoup(htmlcontent,'html.parser')
print(soup)
```

<div style="border-radius:10px; border:black solid; padding: 15px; background-color: purple; font-size:80%; text-align:left">
<p style="font-family:Georgia; font-weight:bold; letter-spacing: 2px; color:white; font-size:200%; text-align:center;padding: 0px;"> Step 4 - To pull the data from html codes efficiently </p>

```python
print(soup.prettify())   
```

<div style="border-radius:10px; border:black solid; padding: 15px; background-color: purple; font-size:80%; text-align:left">
<p style="font-family:Georgia; font-weight:bold; letter-spacing: 2px; color:white; font-size:200%; text-align:center;padding: 0px;"> Step 5 - Taking out the necessary information  </p>

```python
website_content=soup.find_all('a',href=True,class_='_1fQZEK')
print(len(website_content))
for a in website_content:
    name=a.find('div',class_='_4rR01T')
    price=a.find('div',attrs={'class': '_30jeq3 _1_WHN1'})
    rating=a.find('div',attrs={'class': '_3LWZlK'})
    print(name)
```

```python
products=[] # List to store the product name
prices=[] # List to store the product price
ratings=[] # List to store the product rating
```

```python
website_content=soup.find_all('a',href=True,class_='_1fQZEK')
for a in website_content:
    name=a.find('div',class_='_4rR01T')
    price=a.find('div',attrs={'class': '_30jeq3 _1_WHN1'})
    rating=a.find('div',attrs={'class': '_3LWZlK'})
    products.append(name.text)
    prices.append(price.text)
    ratings.append(rating.text)   
```

<div style="border-radius:10px; border:black solid; padding: 15px; background-color: purple; font-size:80%; text-align:left">
<p style="font-family:Georgia; font-weight:bold; letter-spacing: 2px; color:white; font-size:200%; text-align:center;padding: 0px;"> Step 6 - Converting it into dataframe  </p>

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Product name':products,'Price':prices,'Rating':ratings})
print(df)
```
